Uniform Cost Search

```python
class Node:
    def __init__(self, state, parent=None, cost=0):
        self.state = state
        self.parent = parent
        self.cost = cost

    def sort(frontier):
        n = len(frontier)
        for i in range(n):
            min_idx = i
            for j in range(i + 1, n):
                if frontier[j].cost < frontier[min_idx].cost:
                    min_idx = j
            frontier[i], frontier[min_idx] = frontier[min_idx], frontier[i]

class UCS:
    def __init__(self, graph, start, goal):
        self.graph = graph
        self.start = start
        self.goal = goal

    def search(self):
        explored = set()
        frontier = [Node(self.start, None, 0)]

        while frontier:
            Node.sort(frontier)
            node = frontier.pop(0)
            explored.add(node.state)

            print(f"F: {[n.state for n in frontier]}")
            print(f"E: {explored}")
            print(f"Node: {node.state} Parent: {node.parent} Cost: {node.cost}\n")

            if node.state == self.goal:
                print("Goal node found")
                print(f"Path Cost : {node.cost}")
                return

            for neighbour, cost in self.graph[node.state].items():
                if neighbour not in explored:
                    neighbour_node = Node(neighbour, node.state, node.cost + cost)
                    frontier.append(neighbour_node)
        
        print("Goal node not found")

if __name__ == '__main__':
    graph = {
        'a': {'b': 20, 'c': 10},
        'b': {'a': 20, 'c': 50, 'd': 70},
        'c': {'b': 50, 'd': 40, 'e': 30},
        'd': {'b': 70, 'c': 40, 'e': 60},
        'e': {'c': 30, 'd': 60}
    }
    start = 'a'
    goal = 'e'
    ucs = UCS(graph, start, goal)
    ucs.search()
```

Depth-First Iterative Deepening Search

```python
class Node:
    def __init__(self, state, parent, depth):
        self.state = state
        self.parent = parent
        self.depth = depth

class DFIDS:
    def __init__(self, graph, max_depth):
        self.graph = graph
        self.max_depth = max_depth

    def search(self, start, goal):
        for depth in range(self.max_depth + 1):
            explored = []
            frontier = []
            frontier.append(Node(start, None, 0))
            
            while frontier:
                print('f:', [node.state for node in frontier], '\ne:', explored, '\n\n')
                cur_node = frontier.pop()
                print("Node:", cur_node.state, " Parent:", cur_node.parent, " Depth:", cur_node.depth)
                
                if cur_node.state == goal:
                    print("\nGoal node found")
                    return
                
                if cur_node.depth < self.max_depth and cur_node.state not in explored:
                    explored.append(cur_node.state)
                    for neighbor in self.graph[cur_node.state]:
                        if neighbor not in explored:
                            frontier.append(Node(neighbor, cur_node.state, cur_node.depth + 1))
            
            print("Goal node not found within depth limit")

if __name__ == '__main__':
    graph = {'1': ['2', '7', '8'], '2': ['3', '6'], '3': ['4', '5'], '4': [], '5': [], '6': [], '7': [], '8': ['9', '12'],
             '9': ['10', '11'], '10': [], '11': [], '12': []}
    start = '1'
    goal = '6'
    
    dfids = DFIDS(graph, 2)  # Set max_depth to 5
    dfids.search(start, goal)
```

```python
class Graph:
    def __init__(self,graph):
        self.graph=graph
    def reverse(self):
        revers={}
        for node in self.graph:
            revers[node]=[]
        for node,neighbours in self.graph.items():
            for neighbour in neighbours:
                revers[neighbour].append(node)
        return revers
class BDS:
    def __init__(self,graph):
        self.graph=graph
    def Search(self,start,goal):
        rev_graph=Graph(self.graph).reverse()
        frontier1=[[start,None]]
        frontier2=[[goal,None]]
        explored1=[]
        explored2=[]
        path1=[]
        path2=[]
        intersect=None
        while frontier1 and frontier2:
            print("F_front : ",frontier1)
            print("E_front : ",explored1)
            node1,parent1=frontier1.pop()
            path1.append([node1,parent1])
            if node1 in explored2:
                intersect=node1
                break
            if node1 not in explored1:
                explored1.append(node1)
                for neighbour in self.graph[node1]:
                    frontier1.append([neighbour,node1])
            print("F_back : ",frontier2)
            print("E_back : ",explored2)
            print("\n")
            node2,parent2=frontier2.pop()
            path2.append([node2,parent2])
            if node2 in explored1:
                intersect=node2
                break
            if node2 not in explored2:
                explored2.append(node2)
                for neighbour in rev_graph[node2]:
                    frontier2.append([neighbour,node2])
        print("\n")    
        path2=[[node2,node1] for node1,node2 in path2]
        print("Path : ",path1+path2[::-2])
        print("The Search intersected at ",intersect)


graph= {'A':['E'], 'E':['B','G'], 'B':[], 'G':['F','H'], 'F':['C','D'], 'C':[], 'D':[], 'H':['I'], 'I':['J','K'], 
'J':['L','M'], 'L':[], 'M':[], 'K':['N','O'], 'N':[], 'O':[]}
start='A'
goal='O'
bds=BDS(graph)
bds.Search(start,goal)
```

Water jug problem(BFS)

```python
class WaterJug:
    def __init__(self, b, s, t):
        self.b = b
        self.s = s
        self.t = t

    def fbg(self, frontier, explored, big, small):
        if (self.b, small) not in frontier and (self.b, small) not in explored:
            frontier.append((self.b, small))

    def fsg(self, frontier, explored, big, small):
        if (big, self.s) not in frontier and (big, self.s) not in explored:
            frontier.append((big, self.s))

    def ebg(self, frontier, explored, big, small):
        if (0, small) not in frontier and (0, small) not in explored:
            frontier.append((0, small))

    def esg(self, frontier, explored, big, small):
        if (big, 0) not in frontier and (big, 0) not in explored:
            frontier.append((big, 0))

    def pbg(self, frontier, explored, big, small):
        transfer = min(big, self.s - small)
        if (big - transfer, small + transfer) not in frontier and (big - transfer, small + transfer) not in explored:
            frontier.append((big - transfer, small + transfer))

    def psg(self, frontier, explored, big, small):
        transfer = min(small, self.b - big)
        if (big + transfer, small - transfer) not in frontier and (big + transfer, small - transfer) not in explored:
            frontier.append((big + transfer, small - transfer))

    def bfs(self):
        frontier = []
        explored = set()
        path = []
        frontier.append((0, 0))

        while frontier:
            print("F : ", frontier)
            print("E : ", explored,"\n")
            big, small = frontier.pop(0)

            if big == self.t or small == self.t:
                print("Goal Achieved")
                print(path)
                break

            if (big, small) not in explored:
                explored.add((big, small))
                path.append((big, small))
                self.fbg(frontier, explored, big, small)
                self.fsg(frontier, explored, big, small)
                self.ebg(frontier, explored, big, small)
                self.esg(frontier, explored, big, small)
                self.pbg(frontier, explored, big, small)
                self.psg(frontier, explored, big, small)

WaterJug(4, 3, 2).bfs()
```

```python
path = []

class Miscal:
    def __init__(self, lm, lc, rm, rc, boat):
        self.lm = lm
        self.lc = lc
        self.rm = rm
        self.rc = rc
        self.boat = boat
    
    def check(self, a, b, c, d):
        return (a >= 0 and b >= 0 and c >= 0 and d >= 0 and 
                (a == 0 or a >= b) and 
                (c == 0 or c >= d))
    
    def newstate(self, frontier, na, nb, nc, nd, new_bt):
        if self.check(na, nb, nc, nd) and (na, nb, nc, nd, new_bt) not in frontier:
            frontier.append((na, nb, nc, nd, new_bt))
    
    def ronem(self, frontier, a, b, c, d, bo):
        self.newstate(frontier, a - 1, b, c + 1, d, 1)
    
    def ronec(self, frontier, a, b, c, d, bo):
        self.newstate(frontier, a, b - 1, c, d + 1, 1)
    
    def rtwom(self, frontier, a, b, c, d, bo):
        self.newstate(frontier, a - 2, b, c + 2, d, 1)
    
    def rtwoc(self, frontier, a, b, c, d, bo):
        self.newstate(frontier, a, b - 2, c, d + 2, 1)
    
    def ronemonec(self, frontier, a, b, c, d, bo):
        self.newstate(frontier, a - 1, b - 1, c + 1, d + 1, 1)
    
    def lonem(self, frontier, a, b, c, d, bo):
        self.newstate(frontier, a + 1, b, c - 1, d, 0)
    
    def lonec(self, frontier, a, b, c, d, bo):
        self.newstate(frontier, a, b + 1, c, d - 1, 0)
    
    def ltwom(self, frontier, a, b, c, d, bo):
        self.newstate(frontier, a + 2, b, c - 2, d, 0)
    
    def ltwoc(self, frontier, a, b, c, d, bo):
        self.newstate(frontier, a, b + 2, c, d - 2, 0)
    
    def lonemonec(self, frontier, a, b, c, d, bo):
        self.newstate(frontier, a + 1, b + 1, c - 1, d - 1, 0)
    
    def bfs(self):
        explored = set()
        frontier = [(self.lm, self.lc, self.rm, self.rc, self.boat)]
        
        while frontier:
            print('f:', frontier, '\ne:', explored, '\n\n')
            a, b, c, d, bt = frontier.pop(0)
            
            if a == 0 and b == 0 and c == 3 and d == 3 and bt == 1:
                print("Crossed Successfully...\n")
                print(path + [(a, b, c, d, bt)])
                return
            
            if (a, b, c, d, bt) not in explored and (a, b, c, d, bt) not in frontier:
                explored.add((a, b, c, d, bt))
                path.append((a, b, c, d, bt))
                
                if bt == 0:
                    self.ronem(frontier, a, b, c, d, bt)
                    self.ronec(frontier, a, b, c, d, bt)
                    self.rtwom(frontier, a, b, c, d, bt)
                    self.rtwoc(frontier, a, b, c, d, bt)
                    self.ronemonec(frontier, a, b, c, d, bt)
                    
                if bt == 1:
                    self.lonem(frontier, a, b, c, d, bt)
                    self.lonec(frontier, a, b, c, d, bt)
                    self.ltwom(frontier, a, b, c, d, bt)
                    self.ltwoc(frontier, a, b, c, d, bt)
                    self.lonemonec(frontier, a, b, c, d, bt)

lm = 3
lc = 3
rm = 0
rc = 0
boat = 0
Miscal(lm, lc, rm, rc, boat).bfs()
```

```python
class VacWorld:
    def __init__(self, room1, room2, pos):
        self.room1 = room1
        self.room2 = room2
        self.pos = pos

    def sweep(self, frontier, explored, room1, room2, pos):
        if pos == "right" and room2 == "dirty":
            room2 = "clean"
        if pos == "left" and room1 == "dirty":
            room1 = "clean"
        if [room1, room2, pos] not in frontier and [room1, room2, pos] not in explored:
            frontier.append([room1, room2, pos])

    def move_l(self, frontier, explored, room1, room2, pos):
        if pos == 'right':
            pos = 'left'
        if [room1, room2, pos] not in frontier and [room1, room2, pos] not in explored:
            frontier.append([room1, room2, pos])

    def move_r(self, frontier, explored, room1, room2, pos):
        if pos == 'left':
            pos = 'right'
        if [room1, room2, pos] not in frontier and [room1, room2, pos] not in explored:
            frontier.append([room1, room2, pos])

    def bfs(self):
        explored = []
        frontier = [[self.room1, self.room2, self.pos]]
        path = []

        while frontier:
            room1, room2, pos = frontier.pop(0)
            print('f:', frontier, '\ne:', explored, '\n\n')
            
            if room1 == 'clean' and room2 == 'clean':
                path.append([room1, room2, pos])
                print(path)
                return
            
            if [room1, room2, pos] not in explored and [room1, room2, pos] not in frontier:
                explored.append([room1, room2, pos])
                path.append([room1, room2, pos])

            self.sweep(frontier, explored, room1, room2, pos)
            self.move_l(frontier, explored, room1, room2, pos)
            self.move_r(frontier, explored, room1, room2, pos)

room1 = 'dirty'
room2 = 'dirty'
pos = input("Enter position of the vacuum: ")
VacWorld(room1, room2, pos).bfs()
```

Greedy Best First Search

```python
class Node:
    def __init__(self,state,parent,fn):
        self.state=state
        self.parent=parent
        self.fn=fn
class GSearch:
    def __init__(self,graph):
        self.graph=graph
        self.frontier=[]
        self.explored=[]
    def sort(self):
        n=len(self.frontier)
        for i in range(n):
            min=i
            for j in range(i+1,n):
                if self.frontier[j].fn<self.frontier[min].fn:
                    min=j
            self.frontier[i],self.frontier[min]=self.frontier[min],self.frontier[i]
    def best(self,start,goal):
        self.frontier.append(Node(start,None,self.graph[start]['X']))
        while self.frontier:
            
            self.sort()
            node=self.frontier.pop(0)
            
            if node.state==goal:
                print("Goal Acheived")
                break
            if node.state not in self.explored:
                self.explored.append(node.state)
                for neighbour in self.graph[node.state]:
                    if neighbour!='X' and neighbour not in self.explored:
                        fn1=self.graph[neighbour]['X']
                        self.frontier.append(Node(neighbour,node.state,fn1))
graph = eval(input("Enter graph: "))
search = GSearch(graph)
start = input("Enter start node:")
goal = input("Enter goal node:")
search.best(start,goal)
```

```python
def sort_frontier(frontier):
    n = len(frontier)
    for i in range(n):
        for j in range(n):
            if frontier[i][1] < frontier[j][1]:
                frontier[i], frontier[j] = frontier[j], frontier[i]

def heuristic(start, goal):
    count = 0
    for i in range(len(start)):
        if start[i] != goal[i]:
            count += 1
    return count

def up(frontier, explored, state, goal):
    zero = state.index(0)
    if zero not in [0, 1, 2]:
        new_state = state[:]
        new_state[zero], new_state[zero - 3] = new_state[zero - 3], new_state[zero]
        if new_state not in [item[0] for item in frontier] and new_state not in explored:
            frontier.append([new_state, heuristic(new_state, goal)])

def down(frontier, explored, state, goal):
    zero = state.index(0)
    if zero not in [6, 7, 8]:
        new_state = state[:]
        new_state[zero], new_state[zero + 3] = new_state[zero + 3], new_state[zero]
        if new_state not in [item[0] for item in frontier] and new_state not in explored:
            frontier.append([new_state, heuristic(new_state, goal)])

def left(frontier, explored, state, goal):
    zero = state.index(0)
    if zero not in [0, 3, 6]:
        new_state = state[:]
        new_state[zero], new_state[zero - 1] = new_state[zero - 1], new_state[zero]
        if new_state not in [item[0] for item in frontier] and new_state not in explored:
            frontier.append([new_state, heuristic(new_state, goal)])

def right(frontier, explored, state, goal):
    zero = state.index(0)
    if zero not in [2, 5, 8]:
        new_state = state[:]
        new_state[zero], new_state[zero + 1] = new_state[zero + 1], new_state[zero]
        if new_state not in [item[0] for item in frontier] and new_state not in explored:
            frontier.append([new_state, heuristic(new_state, goal)])

def a_star_8queen(start, goal):
    frontier = []
    explored = []
    frontier.append([start, heuristic(start, goal)])
    while frontier:
        print("Frontier: ", frontier)
        print("Explored: ", explored)
        sort_frontier(frontier)
        state, h = frontier.pop(0)
        print("State: ", state, " Heuristic: ", h, '\n\n')
        if state == goal:
            print("Goal state achieved: ", goal)
            return
        if state not in explored:
            explored.append(state)
            up(frontier, explored, state, goal)
            down(frontier, explored, state, goal)
            left(frontier, explored, state, goal)
            right(frontier, explored, state, goal)

if __name__ == '__main__':
    start = eval(input("Enter start state: "))
    goal = eval(input("Enter goal state: "))
    a_star_8queen(start, goal)
```

```python
def Cost(H, condition, weight=1):
    cost = {}
    
    if 'AND' in condition:
        AND_nodes = condition['AND']
        Path_A = ' AND '.join(AND_nodes)
        PathA = sum(H[node] + weight for node in AND_nodes)
        cost[Path_A] = PathA
    
    if 'OR' in condition:
        OR_nodes = condition['OR']
        Path_B = ' OR '.join(OR_nodes)
        PathB = min(H[node] + weight for node in OR_nodes)
        cost[Path_B] = PathB
    
    return cost

def update_cost(H, Conditions, weight=1):
    Main_nodes = list(Conditions.keys())
    Main_nodes.reverse()
    least_cost = {}
    
    for key in Main_nodes:
        condition = Conditions[key]
        print(key, ':', Conditions[key], '>>>', Cost(H, condition, weight))
        c = Cost(H, condition, weight)
        H[key] = min(c.values())
        least_cost[key] = Cost(H, condition, weight)
    
    return least_cost

def shortest_path(Start, Updated_cost, H):
    Path = Start
    
    if Start in Updated_cost.keys():
        Min_cost = min(Updated_cost[Start].values())
        key = list(Updated_cost[Start].keys())
        values = list(Updated_cost[Start].values())
        Index = values.index(Min_cost)
        Next = key[Index].split()
        
        if len(Next) == 1:
            Start = Next[0]
            Path += '<--' + shortest_path(Start, Updated_cost, H)
        else:
            Path += '<--(' + key[Index] + ') '
            Start = Next[0]
            Path += '[' + shortest_path(Start, Updated_cost, H) + ' + '
            Start = Next[-1]
            Path += shortest_path(Start, Updated_cost, H) + ']'
    
    return Path

H = eval(input('Enter nodes with heuristic costs: '))
Conditions = eval(input('Enter graph: ')) 
weight = 1
print('Updated Cost:')
Updated_cost = update_cost(H, Conditions, weight=1)
print('Shortest Path:\n', shortest_path('A', Updated_cost, H))

"""Enter nodes with heuristic costs: {'A':-1,'B':5,'C':2,'D':4,'E':7,'F':9,'G':3,'H':0 ,'I':0,'J':0} 
Enter graph: {'A':{'OR':['B'],'AND':['C','D']},'B':{'OR':['E','F']},'C':{'OR':['G'] 
,'AND':['H','I']},'D':{'OR':['J']}}
"""
```

MINI MAX

```python
import math

def minimax(game, state):
    def maxvalue(game, state):
        if state not in game:
            return state
        
        v = -math.inf
        for a in game[state]:
            v = max(v, minvalue(game, a))
        
        print("Max value for ", state, " : ", v)
        return v
    
    def minvalue(game, state):
        if state not in game:
            return state
        
        v = math.inf
        for a in game[state]:
            v = min(v, maxvalue(game, a))
        
        print("Min value for ", state, " : ", v)
        return v
    
    return max([minvalue(game, a) for a in game[state]])

graph = eval(input("Enter graph: "))
ans = minimax(graph, 'A')
print("The optimal value is: ", ans)

# {'A':['B','C'],'B':['D','E'],'C':['F','G'],'D':[-1,4],'E':[2,6],'F':[-3,-5],'G':[0,7]}
```

```python
import math

def alphaBeta(game, state):
    def maxvalue(game, state, alpha, beta):
        if state not in game:
            return state
        
        v = -math.inf
        for a in game[state]:
            v = max(v, minvalue(game, a, alpha, beta))
            if v >= beta:
                print("Pruned subtree at state:", state, "with value:", v)
                return v
            alpha = max(alpha, v)
            print("State: ", state, " V: ", v, " Alpha: ", alpha, " Beta: ", beta)
        
        return v
    
    def minvalue(game, state, alpha, beta):
        if state not in game:
            return state
        
        v = math.inf
        for a in game[state]:
            v = min(v, maxvalue(game, a, alpha, beta))
            if v <= alpha:
                print("Pruned subtree at state:", state, "with value:", v)
                return v
            beta = min(beta, v)
            print("State: ", state, " V: ", v, " Alpha: ", alpha, " Beta: ", beta)
        
        return v
    
    alpha = -math.inf
    beta = math.inf
    v = maxvalue(game, state, alpha, beta)
    
    return v

graph = eval(input("Enter graph: "))
ans = alphaBeta(graph, 'A')
print("The optimal value is: ", ans)
# {'A':['B','C'],'B':['D','E'],'C':['F','G'],'D':[1,4],'E':[7,2],'F':[3,0],'G':[6,5]}
```

```python
import itertools

def number(n, d):
    t = 0
    for i in n:
        t = d[i] + (t * 10)
    return t

def test(l, s, d):
    total = 0
    for i in l:
        total += number(i, d)
    return total == number(s, d)

def check(d, c):
    for i in d.keys():
        if i in c and d[i] == 0:
            return True
    return False

def solve(l, s):
    c = [i[0] for i in l] + [s[0]]
    p = list(set(''.join(l) + s))
    q = len(p)
    permutations = itertools.permutations(range(0, 10), q)
    
    for perm in permutations:
        d = {p[j]: perm[j] for j in range(q)}
        
        if check(d, c):
            continue
        
        if test(l, s, d):
            return d
    
    return None

if __name__ == "__main__":
    l = input('Enter list of strings: ').split()
    s = input('Enter output string: ')
    
    solution = solve(l, s)
    
    if solution:
        print(solution)
        print('Solution found')
    else:
        print('No solution found')

#TWO TWO
#FOUR
```
